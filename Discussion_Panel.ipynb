# рџ§¬Drug Repurposing Similarity Roundtable

Before building a Drug Similarity Score (DSS), I convened a small virtual advisory panel of computational biology specialists to help determine what evidence actually matters when comparing drugs for repurposing.

I asked three domain experts to debate the strongest signals:

рџ§Є Structural Bioinformatician вЂ” argues molecular similarity should anchor the analysis (Tanimoto coefficient)

рџ§  Network Topologist вЂ” focuses on shared targets and pathway overlap (proteinвЂ“protein interaction networks)

рџ“Љ Systems Biologist вЂ” prioritizes functional evidence from gene-expression reversal (transcriptomics)

Together, they evaluated the strengths and limitations of each signal and reached consensus on a weighted similarity formula designed to balance chemistry, mechanism, and biological effect.

This roundtable forms the foundation of the Gold Layer Drug Similarity Score (DSS) used throughout this project

```python
from crewai import Agent, Task, Crew, Process
from dotenv import load_dotenv
from openai import OpenAI
import json 
import re
from agents import (network_specialist, structural_specialist, systems_specialist)

load_dotenv(override=True)

# 2. Define the Roundtable Task
similarity_task = Task(
    description=f"""
    Discuss and agree on a 'Simple but Meaningful' set of metrics to calculate drug similarity 
    for repurposing. You must debate the pros and cons of:
    1. Tanimoto Coefficients (Structure)
    2. Target Overlap (Network)
    3. Gene Signature Reversal (Transcriptomics)
    Reach a final consensus on how to weigh these three for a 'Gold Layer' similarity score. 
    Return ONLY valid JSON in this exact format:

    {{
      "weights": {{
        "structure": float,
        "network": float,
        "transcriptomics": float
      }},
      "explanation": {{
        "structure": "string",
        "network": "string",
        "transcriptomics": "string"
      }},
      "final_formula": "string"
    }}

    Requirements:
    - weights must sum to exactly 1.0
    - explanation must briefly justify each weight
    - final_formula must be concise and reusable
    """,
    expected_output="JSON with weights, explanation, and final formula",
    agent=systems_specialist
)

# 3. Create the Roundtable (The Crew)
drug_discovery_team = Crew(
    agents=[network_specialist, structural_specialist, systems_specialist],
    tasks=[similarity_task],
    process=Process.sequential, # This forces them to talk and agree
    verbose=True
)

# 4. Kickoff the Discussion
formula_consensus = drug_discovery_team.kickoff()

# 5. Extract JSON safely
json_text = re.search(r"\{.*\}", str(formula_consensus), re.DOTALL).group()
formula_data = json.loads(json_text)
```
